```python
import bs4
import urllib2
import re
```

```python
addr = 'http://paragon.wiki/Heroes'
html = urllib2.urlopen(addr).read()
text = bs4.BeautifulSoup(html, 'lxml').get_text()
```

```python
print text
```

```python
re.findall(r'(?<=\n\n\n\n)([\w. ]+)(?:\n\n\n\n)', text)
```

```python
addr = 'http://paragon.wiki/Sparrow'
html = urllib2.urlopen(addr).read()
text = bs4.BeautifulSoup(html, 'lxml').get_text()
```

```python
re.search(r'(?:Physical|Energy)\s+Damage: (\d+)', text).group(1)
```

```python
re.search(r'Attack Speed Level Bonus:\s+\(\+([\d.]+)%\)', text, re.M).group(1)
```

```python
re.search(r'Damage Increase per Level:\s+([\d.]+)', text).group(1)
```
